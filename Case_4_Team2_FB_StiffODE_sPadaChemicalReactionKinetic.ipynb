## **SOLUSI**

```python
# Persamaan sesuai problem 28.12
def reaction_kinetics(t, c):
    c1, c2, c3 = c
    dc1_dt = -0.013 * c1 - 1000 * c1 * c3
    dc2_dt = -2500 * c2 * c3
    dc3_dt = -0.013 * c1 - 1000 * c1 * c3 - 2500 * c2 * c3
    return np.array([dc1_dt, dc2_dt, dc3_dt])

# Metode Backward Euler
def Backward_Euler(x):
    BDF = ci + h * reaction_kinetics(nilai_t[i], x)
    return x - BDF

# Import Library yang digunakan
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from scipy.optimize import root

# Anggota kelompok
print('Kelompok 2 - Fisika B')
print('1. Dhinar Hidayath  		(1306621054)')
print('2. Khansa Farras Callista 	(1306621067)')
print('3. Puspa Zuhrotul Kamilah 	(1306621058)')
print('4. Salma Mardhiyah 	    	(1306621061)\n')
print("="*100, "Studi Kasus Stiff ODE pada Kinetika Reaksi Kimia " \
      "Menggunakan Backward Euler dan Implicit Runge Kutta", "="*100, "", sep='\n')

# Inisiasi waktu yang dibutuhkan
t0 = 0                              # waktu awal
tf = 50                             # waktu akhir
h  = 0.01                           # step size
n  = int((tf - t0 + h) / h)         # banyak data yang dicari
nilai_t = np.linspace(t0, tf, n - 1)  # rentang waktu t

# Kondisi awal konsentrasi saat t = 0
c0 = np.array([1, 1, 0])

# Matriks nilai konsentrasi
nilai_c = np.zeros((3, n - 1))
nilai_c[:, 0] = c0

# Program utama
for i in range(n - 2):
    ti = nilai_t[i]                 # waktu ke - i
    ci = nilai_c[:, i]              # mengganti nilai_c dengan ci

    # Melakukan Perhitungan Euler eksplicit sebelum menggunakan BDF
    if i < 1:
        c_ni = ci + h * reaction_kinetics(ti, ci)

    # Melakukan Perhitungan Euler implicit (Backward Euler)
    else:
        konsentrasi = root(Backward_Euler, ci)
        c_ni = konsentrasi.x

    konsentrasi = root(Backward_Euler, c_ni)
    nilai_c[:, i + 1] = konsentrasi.x

# Grafik metode Backward Euler
plt.plot(nilai_t, nilai_c[0], label='c1 (BDF)')
plt.plot(nilai_t, nilai_c[1], label='c2 (BDF)')
plt.plot(nilai_t, nilai_c[2], label='c3 (BDF)')

# Grafik dengan solver
RK23 = solve_ivp(reaction_kinetics, (t0, tf), c0, method='Radau')
plt.plot(RK23.t, RK23.y[0], 'r--', linewidth=2, dashes=(5, 5),  label='c1 (Radau)')
plt.plot(RK23.t, RK23.y[1], 'k--', linewidth=2, dashes=(5, 5),  label='c2 (Radau)')
plt.plot(RK23.t, RK23.y[2], 'm--', linewidth=2, dashes=(5, 5),  label='c3 (Radau)')
plt.xlabel('Waktu')
plt.ylabel('Konsentrasi')
plt.title('Persamaan Diferensial Biasa dengan Metode Backward Euler')
plt.legend(loc='best', bbox_to_anchor=(0, 0, 0.3, 0.5))
plt.grid(True)
plt.show()

print("\nNilai c saat t = 50 dengan metode BDF dan Solver")
print(f"- c1 = {nilai_c[0][-1]:.10f}  dan {RK23.y[0][-1]:.10f}")
print(f"- c2 = {nilai_c[1][-1]:.10f}  dan {RK23.y[1][-1]:.10f}")
print(f"- c3 = {nilai_c[2][-1]:.10f} dan {RK23.y[2][-1]:.10f}")
```
